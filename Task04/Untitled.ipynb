```python
year= pd.DataFrame(df.groupby(["iyear"]).count())
```

```python
plt.figure(figsize=(10, 10))
plt.title("Terriosit Activity in last few years")
plt.xlabel("Events")
plt.ylabel("Year")
plt.grid()
plt.plot(year.index, year, marker="D")
plt.show()
```

```python
month= pd.DataFrame(df.groupby(["imonth"]).count())
```

```python
plt.figure(figsize=(10, 10))
plt.title("Terriosit Activity in last few years")
plt.xlabel("Month")
plt.ylabel("Acivity")
plt.grid()
plt.xticks(month.index[1:], calendar.month_name[1:])
plt.plot(month.index[1:], month[1:])
plt.show()
```

```python
obj=df.groupby(["country_txt"])
obj = pd.DataFrame(obj.count())["eventid"]
```

```python
plt.figure(figsize=(40,10))
plt.xticks(rotation=90)
plt.bar(df["country_txt"].unique(), obj)
```

```python
hot_activity=obj[obj>obj.mean()]
```

```python
plt.figure(figsize=(10,10))
plt.xticks(rotation=90)
plt.grid()
plt.xlabel("Country")
plt.ylabel("Event count")
plt.bar(hot_activity.index, hot_activity)
```
