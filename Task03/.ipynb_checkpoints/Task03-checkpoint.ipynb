```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("SampleSuperstore.csv")
```

```python
df.head()
```

```python
dic=dict()
[df[i].unique() for i in df.columns]
```

```python
df.describe()
```

```python
df.shape
```

Droping the country column as it has only unique value in it

```python
df.drop(["Country"],axis=1, inplace=True)
```

Sales from differnt Shipping modes

```python
plt.xlabel("Ship Mode")
plt.ylabel("Sales(Unit)")
plt.title("Sales from different Ship Mode")
plt.bar(df["Ship Mode"].unique(),[df["Sales"][df["Ship Mode"]==i].sum() for i in df["Ship Mode"].unique()],label="Sales")
plt.bar(df["Ship Mode"].unique(),[df["Profit"][df["Ship Mode"]==i].sum() for i in df["Ship Mode"].unique()],label="Profit")
plt.legend()
plt.show()
```

Maximum sales from Standard Class

Sales from different states and their profits

```python
plt.figure(figsize=(10,10))
plt.xlabel("States")
plt.ylabel("Revenue")
plt.xticks(rotation=90)
plt.title("Sales and profit from different states")
plt.bar(df["State"].unique(), [df["Sales"][df["State"]==i].sum() for i in df["State"].unique()],label="Sales")
plt.bar(df["State"].unique(), [df["Profit"][df["State"]==i].sum() for i in df["State"].unique()],label="Profit")
plt.grid()
plt.legend()
plt.show()
```

Maximum sales from califonia and maximum profit too.
Loss in profit maximum is in Texas

Profit to sales ratio

```python
plt.figure(figsize=(10,10))
plt.xlabel("States")
plt.ylabel("Revenue")
plt.xticks(rotation=90)
plt.title("Sales to profit ratio")
plt.bar(df["State"].unique(), np.array([df["Profit"][df["State"]==i].sum() for i in df["State"].unique()])/np.array([df["Sales"][df["State"]==i].sum() for i in df["State"].unique()]),label="Sales")
plt.grid()
plt.legend()
plt.show()
```

It is observed that District of Columnbia has the maximum profit to sales ratio.  Ohio has the maximum loss to sales ratio.

Sales done by various segments

```python
plt.pie([df["Segment"][df["Segment"]==i].count() for i in df["Segment"].unique()], autopct="%1.2f%%")
plt.legend(labels=df["Segment"].unique(), title="The sales in various segments", loc="upper right", bbox_to_anchor=(1.65, 1))
plt.show()
```

Consumer has the maximum purchase.

Sales in each category

```python
plt.pie([sum(df["Quantity"][df["Category"]==i]) for i in df["Category"].unique()], autopct="%1.3f%%")
plt.legend(title="Sales in each category", labels=df["Category"].unique(), loc="upper right", bbox_to_anchor=(1.5, 1))
plt.show()
```

Office supplies are the most sold products category.

Various Categories and their Sub products

```python
for i in df["Category"].unique():
    data=np.array([sum(df["Quantity"][df["Sub-Category"]==j][df["Category"]==i]) for j in df["Sub-Category"].unique()])
    data=data[data>0]
    plt.pie(data, autopct="%1.1f%%")
    plt.legend(labels=df["Sub-Category"].unique(), title=i, loc="center right", bbox_to_anchor=(1.5, 0.5))
    plt.show()
```

Relation between Sales, profit and discount

```python
plt.figure(figsize=(10,10))
plt.title("Relation between Sales, Profit and Discount")
plt.xlabel("Discount")
plt.ylabel("Sales and Profit")
plt.grid()
plt.scatter(df["Discount"], df["Sales"], marker="D")
plt.scatter(df["Discount"], df["Profit"], marker="*")
plt.show()
```

As discount increases the profit decrese.

Loss in each state

```python
profit_neg=df[df["Profit"]<0]
```

```python
state=profit_neg.groupby(["State"])
state_profit=state["Profit"].sum()
state_discount=state["Discount"]
```

```python
state_profit
```

```python
plt.figure(figsize=(8,8))
plt.pie(-state_profit, autopct="%1.2f%%")
plt.legend(state_profit.index, bbox_to_anchor=(1,1.5))
```

Loss in each state(bar graph)

```python
plt.xticks(rotation=90)
plt.bar(state_profit.index, state_profit)
plt.grid()
```

Average discount provided in the states where loss was encountred.

```python
state_disavg=state_discount.sum()/state_discount.count()
```

```python
plt.xticks(rotation=90)
plt.bar(state_disavg.index, state_disavg)
plt.show()
```

Its recommended that the owner has to focus on reducing the losses in the specific states by focuing on reducing the amount discount given by the company on various occassion.

To make more profit the company has to do more sales in country with more profit to sales ratio.
