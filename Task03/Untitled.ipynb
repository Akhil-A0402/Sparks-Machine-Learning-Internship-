# Exploratory Data Analysis - Retail

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("SampleSuperstore.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(["Country"],axis=1, inplace=True)
```

```python
plt.xlabel("Ship Mode")
plt.ylabel("Sales(Unit)")
plt.title("Sales from different Ship Mode")
plt.bar(df["Ship Mode"].unique(),[df["Sales"][df["Ship Mode"]==i].sum() for i in df["Ship Mode"].unique()],label="Sales")
plt.bar(df["Ship Mode"].unique(),[df["Profit"][df["Ship Mode"]==i].sum() for i in df["Ship Mode"].unique()],label="Profit")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.xlabel("States")
plt.ylabel("Revenue")
plt.xticks(rotation=90)
plt.title("Sales and profit from different states")
plt.plot(df["State"].unique(), [df["Sales"][df["State"]==i].sum() for i in df["State"].unique()],label="Sales")
plt.plot(df["State"].unique(), [df["Profit"][df["State"]==i].sum() for i in df["State"].unique()],label="Profit")
plt.grid()
plt.legend()
plt.show()
```

```python
plt.xlabel("Segment")
plt.ylabel("Number of purchases")
plt.title("Total number of purchases in each segment")
plt.bar(df["Segment"].unique(),[df["Segment"][df["Segment"]==i].count() for i in df["Segment"].unique()])
plt.show()
```
